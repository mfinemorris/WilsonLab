```python
#LOAD FILES
#*************************************************
#load the .csv files and store their contents in a dictionary 
#where keys are machine names and values are the contents of
#the .csv files stored in a DataFrame.
def loadFiles():
    
    #path to the file that contains the rate lookup table
    #when this code is in the folder with the lookup tables set path = ""
    path = os.path.expanduser("~/Desktop/Wilson_Lab/AIM_calc/")
    
    #get the files that contains the rate lookup tables
    fileExt = ".csv"
    destdir = path
    #retrieve every .csv file from the directory located at path destdir
    files = [ f for f in os.listdir(destdir) if (os.path.isfile(os.path.join(destdir,f)) and f[-4:] == fileExt) ]
    
    #make a dict with the file names and data
    allServices = {}
    for f in files:
        allServices[f[:len(f)-len(fileExt)]] = pd.DataFrame.from_csv(os.path.join(destdir, f), index_col=0)
    print "Files loaded. Availiable machines are: {}.".format(str(allServices.keys()).strip("[]"))
    return allServices

    
#will load the files and display the avaliable machines by their key names
allServices = loadFiles()
#get the group types
#for key,val in allServices.iteritems(): print "{}: \n {}\n".format(key, val.to_string())
#colNames = [col for col in next(allServices.itervalues()).columns]
#print next(allServices.itervalues()).columns[0]
#print next(allServices.itervalues()).index
```

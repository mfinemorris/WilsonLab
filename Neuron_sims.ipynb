```python
def integrateFire(tau = 1, Vt = -35, Vr = -60, El = -50, connectV = 10, simtime = 100, spikeTriggerTime  = 20):

    reinit_default_clock()
    
    tauUnits = tau * ms
    VtUnits = Vt * mvolt          # spike threshold
    VrUnits = Vr * mvolt          # reset value
    ElUnits = El * mvolt          # resting potential (same as the reset)
    
    eqs = Equations('''
    dV/dt = -(V-ElUnits)/tauUnits : volt
    ''')
    
    #generateGroup generates spike that triggers the spiking of spikeGroup
    generatorGroup = SpikeGeneratorGroup(1, [(0, spikeTriggerTime * ms)])
    spikeGroup = NeuronGroup(N=1, model=eqs, threshold=VtUnits, reset=VrUnits)
    #spikeGroup.V = Vr + rand(1) * (Vt - Vr) #wow, this does wierd stuff

    #connect generator to spikeGroup
    C = Connection(generatorGroup, spikeGroup, 'V')
    C[0,0] = connectV * mvolt
    
    #monitor the voltage of spike group while running simulation
    M = StateMonitor(spikeGroup, 'V', record=0)
    run(simtime * ms)
    
    #plot the voltage as a function of time
    labelString = "tau: {}, Vt: {}, Vr: {}, El: {}, connectV: {}, spikeTriggerTime: {}".format(tau, Vt, Vr, El, connectV, spikeTriggerTime)
    actionPotential = plt.plot(M.times / ms, M[0] / mV, label = labelString )
    xlabel('Time (ms)')
    ylabel('Membrane potential (mV)')
    title('Membrane potential for neuron 0')
    return actionPotential
```

```python
from brian import *
from brian.library.ionic_currents import *
import matplotlib.pyplot as plt
%matplotlib inline

#TODO: Modify to model a calcium channel if necessary (started but not finished)
def Ca_current_HH(gmax, ECa, current_name=None):
    '''
    Hodkin-Huxley Ca+ current.
    '''
    current_name = current_name or unique_id()
    return Current('''
    I=gmax*m**3*h*(ECa-vm) : amp
    dm/dt=alpham*(1-m)-betam*m : 1
    dh/dt=alphah*(1-h)-betah*h : 1
    alpham=.1*(25*mV-vm)/(exp(2.5-.1*vm/mV)-1)/mV/ms : Hz
    betam=4*exp(-.0556*vm/mV)/ms : Hz
    alphah=.07*exp(-.05*vm/mV)/ms : Hz
    betah=1./(1+exp(3.-.1*vm/mV))/ms : Hz
    ''', gmax=gmax, ENa=ENa, I=current_name, current_name=current_name)



# a single neuron, obeying a Hodgkins-Huxley model
class HHmodel:
    #TODO: add ability for user to specify the capacitance/area (g_leak, g_Na, g_K)
    
    # create a neuron that obeys a hodgkin-huxley model and stores behavior (in self.runs) of 
    # the membrane voltage (vm) each time the simulation runs.
    # Arguments ending in K refer to potassium, Na to sodium, and l to leak
    # Args starting with E are the reversal voltages (millivolts), 
    # and those starting with g are the conductance per unit area (msiemens)
    # N is the number of neurons initialized in the NeuronGroup, C is the membrane capacitance (units uF)
    #20-50 uF
    def __init__(self, N = 1, El = 10.6, EK = -12.0, ENa = 120.0, C = 1, gl = .3, gK = 36, gNa = 120):
        
        #reinitalize the clock so that the simulation starts at time = 0 (this may be removed or modified later)
        self.simClock = Clock(dt = .01 * ms, t = 0.0 * ms)
        
        #number of cells in simulation
        self.N = 1
        
        # Reversal voltages for each of the channels and the leak
        self.El = El * mV #leak
        self.EK = EK * mV #Potassium
        self.ENa = ENa * mV #Sodium
        #capacitance
        self.C = C * uF
        #conductance per unit area
        self.gK = gK * msiemens #Potassium
        self.gNa = gNa * msiemens #Sodium
        self.gl = gl * msiemens #leak
        
        # construct the equation for dvm/dt based on currents
        self.eqs = MembraneEquation(self.C) + leak_current(self.gl, self.El)
        # add the equations for potassium and sodium channel currents
        self.eqs += K_current_HH(self.gK, self.EK) + Na_current_HH(self.gNa, self.ENa)
        self.eqs += Current('I:amp') #specifies the units for current
        
        # initialize the N neurons
        self.cells = NeuronGroup(self.N, self.eqs, implicit=True, freeze=True, clock = self.simClock, order = 2)
        self.cells.vm = -65.0 * mvolt
        # holds the StateMonitors for all the runs. Basically remembers the voltage history for the cells.
        self.runs = []
    
    # run the simulation starting from the end of the last simulation run
    # store the newest simulation run
    def runSimulator(self, time):
        
        # make a monitor to track the evolution of the membrane voltage (vm) of the cells with time
        # and put the cells and monitor in a network. Run the network to run the simulation
        # the network isn't absolutely necessary, I could just say 'run(time * ms)'
        # but I dislike the idea of a 'current active NeuronGroup'
        M = StateMonitor(self.cells, 'vm', record = True, clock = self.simClock)
        n = Network(self.cells, M)
        n.run(time * ms)
        
        #add the newest run to the runs list
        self.runs.append(M)
        return M
  
    #plot each of the different runs on the same plot, 
    #but with different line styles (line styles repeat every three runs)
    #return the plot
    def plotEachRun(self, whichNeuron = 0):
        if whichNeuron >= self.N or whichNeuron < 0:
            raise IndexError("No neuron exists at index {}.".format(whichNeuron))
        
        runHistory = None
        lineStyle = ['-', '--', ':']
        lineNum = 0
        for state in self.runs:
            # for each state monitor
            # plot the time in milliseconds and the voltage of neuron number whichNeuron in millivolts
            # third arg cycles through the elements of lineStyle
            runHistory = plot(state.times/ms, state[whichNeuron]/mV, lineStyle[lineNum%len(lineStyle)])
            lineNum+=1
        xlabel('Time (ms)')
        ylabel('Membrane potential (mV)')
        title('Membrane potential for neuron '+ str(whichNeuron))
        return runHistory    
    
    
    #plot each of the states as one combined line which is generated from self.runs
    #does essentially the same thing as plotCombinedRuns()
    #return the plot
    def plotVoltageHistory(self, whichNeuron = 0):
        #check that the specified neuron exists
        if whichNeuron >= self.N or whichNeuron < 0:
            raise IndexError("No neuron exists at index {}.".format(whichNeuron))
            
        # collect the time and voltage data from different runs of the simulator 
        # and concatenate them to make a single, continuous curve
        times = concatenate([run.times for run in self.runs],0)
        voltages = concatenate([run[whichNeuron] for run in self.runs],0)
        voltageHistory = plot(times/ms, voltages/mV)
        xlabel('Time (ms)')
        ylabel('Membrane potential (mV)')
        title('Membrane potential for neuron '+ str(whichNeuron))
        return voltageHistory
    
    # print the values for the various user-defined voltages and the number of neurons in the group
    def __repr__(self):
        return "N:{} cells, E_leak: {} mV, E_K: {} mV, E_Na: {} mV".format(self.N, self.El/mV, self.EK/mV, self.ENa/mV)
    
#initalize the neuron
singleNeuron = HHmodel()

#run the neuron for 50 milliseconds
singleNeuron.runSimulator(50)

#add a driving current (stimulus?) then follow for 50 mseconds
singleNeuron.cells.I = 10 * uA
singleNeuron.runSimulator(50)

#lower driving current and run another 50 seconds
singleNeuron.cells.I = 5 * uA
singleNeuron.runSimulator(50)

print "Plot 1"
# plot each run with a different line style and color
singleNeuron.plotEachRun()
#plt.show()

print "Plot 2"
# plot all the runs as a single line
singleNeuron.plotVoltageHistory()

#plt.show()
```

````python
import scipy.integrate as spi
import numpy as np
import matplotlib.pyplot as plt
import warnings
import pandas as pd
import math
import inspect
import copy
from time import clock
import sys

#globals().update(the_dict)
folderPath = '/Users/morganfine-morris/Dropbox/Wilson_lab_papers/Morgan/Voltages/'

# Do NOT move this. It is here because it is needed in the definition of df_atp to determine whether to store the component
# currents in global arrays
collect_current_data = False

# Time each run of the integrator
timing_on = False

#------------------------------------------------------------------------------------------
#### INITIAL CONDITIONS ####
Vs0=-60.0 #initial cell memberane voltage
ns0=0.004 #initial value of gating var n
hs0=0.33 #initial value of gating var h
Vd0=-50.0 #initial dendritic membrane voltage
C0=0.03 #initial calcium 2+ balance
l0=0.93 # initial value of IP3 channel gating variable

'''
Initial states of the ATP channels. Numerical values represent the fraction of the cell's P2X7 receptors in that state.
They must sum to 1. If a state = 0, then none of the P2X7 receptors are in that state. 
If a state = 1.0, then all of the P2X7 receptors are in that state, and the remaining states must all = 0.

state name : state attributes
q1 : double-liganded, open, unsensitized
q2 : triple-liganded, open, unsensitized
q3 : triple-liganded, open, sensitized
q4 : double-liganded, open, sensitized
c1 : unliganded, closed, unsensitized
c2 : mono-liganded, closed, unsensitized
c3 : mono-liganded, closed, sensitized
c4 : unliganded, closed, sensitized
'''
q1o = 0.
q2o = 0.
q3o = 0.
q4o = 0.
c1o = 1.0
c2o = 0.
c3o = 0.
c4o = 0.


#Initial atp concentration (units of Molars)
atp0=3.2e-6 #Molars

#### TIME (ms) ####
tStart=0.
tEnd=1000.*60#*3.
dt=1.

#### MEMBRANE CAPACITANCE (pF) ####
Cms=21
#dendrite membrane capacitance
Cmd=5

#### REVERSAL POTENTIALS (mV) ####
# sodium potential (for normal sodium current and persistent sodium current)
ena=50
# potassium potential
ek=-85
# leak reversal potential
eL=-54.0 # suggested value for vleak from Butera paper are -60,-57.5,-54
# reversal potential of non-NMDA glutamatergic synaptic currents.
eSyn=0.0

#### CONDUCTANCE (nS) ####
# potassium current conductance
gk=11.2
# sodium current conductance
gna=28
# persistent sodium current conductance
gnaps=2.8
# leak current conductance
gL=5.6
# tonic conductance
gtonic=0.0
# calcium channel conductance
gcan = 1.0

#### APPLIED CURRENT ####
Iaps=0

#### HALF (IN)ACTIVATION VOLTAGE (mV) ####
vm=-34
vn=-29
vmp=-40
vh=-48

#### INVERSE SLOPE MULTIPLIER (mV) ####
sm=-5
sn=-4
smp=-6
sh=6

#### TIME CONSTANT MAX VALUES (ms) #####
taunb=10 #tau max for n
tauhb=10000 #tau max for h

#### Constants for calculating Ca Flux: ER --> Cytosol ####
I=1 #IP3 concentration
LL=0.37 #ER leak permeability
P=31000 #maximum total permeability of IP3 channels
Ki=1.0 #dissociation consts for IP3 receptor activity by IP3
Ka=0.4 #dissociation consts for IP3 receptor activity by Ca

#### Constants for calculating Ca Flux: Cytosol --> ER ####
Ve=400 #Maximal SERCA pump rate
Ke=0.2 #coefficient for SERCA pumps

#### ER Ca CONCENTRATION ####
Ct=1.25 #Total Ca
sigma=0.185 #ratio of cytosolic to ER volume

#The ER parameters
fi=0.0001 #bound Ca concentration in cytosol
Vi=4 #free Ca concentration in cytosol
A=0.005 #scaling const.
Kd=0.4 #dissociation constant for IP3 receptor inactivation by Ca

#### Calcium current activation ####
Kcan=0.74 # microM 
ncan=0.97

#### Ratio of somatic to total area ####
k=0.3

#### P2X7 RECEPTOR ####
# RATE CONSTANTS
# Back rates (ms)**-1
k1 = 0.3e-3 
k3 = 2.4e-3
k5 = 1.58e-3
# Forward rates (ms*M)**-1
k2 = 40000e-3
k4 = 50000.0e-3
k6 = 7000e-3
# Sensitized/unsensitized rates
L1 = 0.0001e-3
L2 = 0.004e-3
L3 = 0.5e-3

# CONDUCTANCE (nS)
#conductance of unsensitized channels
g12 = 15 # nS
#conductance of sensitized channels
g34 = 45 # nS

#P2X7 channel reversal potential
eatp = 0.0 #mV

param = {'Cms': Cms,  'Cmd': Cmd,  'ena': ena,  'ek': ek,  'eL': eL,  'eSyn': eSyn,  'gk': gk,\
        'gna': gna,  'gnaps': gnaps,  'gL': gL,  'gtonic': gtonic,  'gcan': gcan ,  'Iaps': Iaps,  'vm': vm,\
        'vn': vn,  'vmp': vmp,  'vh': vh,  'sm': sm,  'sn': sn,  'smp': smp,  'sh': sh,  'taunb': taunb,\
        'tauhb': tauhb,  'I': I,  'LL': LL,  'P': P,  'Ki': Ki,  'Ka': Ka,  'Ve': Ve,  'Ke': Ke,  'Ct': Ct,\
        'sigma': sigma,  'fi': fi,  'Vi': Vi,  'A': A,  'Kd': Kd,  'Kcan': Kcan,  'ncan': ncan,  'k': k, \
        'k1': k1, 'k2': k2, 'k3': k3, 'k4': k4, 'k5': k5, 'k6': k6, 'L1': L1, 'L2': L2, 'L3': L3, 'g12': g12,\
        'g34': g34, 'eatp': eatp, 'Vs0': Vs0 , 'ns0': ns0 , 'hs0': hs0 , 'Vd0': Vd0 , 'C0': C0 , 'l0': l0 , \
        'q1o': q1o , 'q2o': q2o , 'q3o': q3o , 'q4o': q4o , 'c1o': c1o , 'c2o': c2o , 'c3o': c3o , 'c4o': c4o , 'atp0': atp0}

atp_current = []
ca_current = []
nap_current = []
na_current = []
k_current = []
timeArray = []
def df_atp(y, t, atpFunct, Iaps , gnaps , gk , vmp , gL , Ka , Ke , Kd , Ki , \
                                        Cmd , eSyn , P , Cms , taunb , g34 , gna , Kcan , Ct , ek , vh , \
                                        vm , vn , gtonic , eL , LL , tauhb , smp , k3 , k2 , k1 , k6 , k5 , \
                                        k4 , A , gcan , ena , I , L2 , L3 , L1 , fi , ncan , Ve , Vi , \
                                        k , eatp , sh , sn , sm , sigma , g12):
    
    Vs = y[0]
    ns = y[1]
    hs = y[2]
    Vd = y[3] #Dendritic membrane voltage
    C = y[4] #Intracellular Ca2+ balance
    l = y[5] #IP3 channel gating variable
    q1 = y[6]
    q2 = y[7]
    q3 = y[8]
    q4 = y[9]
    c1 = y[10]
    c2 = y[11]
    c3 = y[12]
    c4 = y[13]
    atp = y[14]


    # SOMATIC FUNCTIONS
    minfs = 1/(1+math.exp((Vs-vm) /sm))
    ninfs = 1/(1+math.exp((Vs-vn) /sn))
    minfps = 1/(1+math.exp((Vs-vmp)/smp))
    hinfs = 1/(1+math.exp((Vs-vh) /sh))

    tauns = taunb/math.cosh((Vs-vn)/(2*sn))
    tauhs = tauhb/math.cosh((Vs-vh)/(2*sh))

    # DENDRITIC FUNCTIONS
    #Calculate ER Ca
    Ce = (Ct - C)/sigma
    # Flux of Ca from ER to cytosol(regulated by IP3 receptors)
    J_ER_in=(LL + P*math.pow(((I*C*l)/((I+Ki)*(C+Ka))),3))*(Ce - C) 
    # Flux from cytosol back to ER (controlled by SERCA pumps)
    J_ER_out=Ve*math.pow(C,2)/(math.pow(Ke,2)+math.pow(C,2))
    # Activation of calcium current (I_can)
    caninf = 1/(1+math.pow((Kcan/C),ncan))

    # CURRENT EXPRESSIONS
    # currents in Soma
    I_nas = gna*math.pow(minfs,3)*(1-ns)*(Vs-ena) #sodium current
    I_ks = gk*math.pow(ns,4)*(Vs-ek) #potassium current
    I_naps = gnaps*minfps*hs*(Vs-ena) #persistent sodium current
    I_L = gL*(Vs-eL) #pA
    I_atp = (g12*(q1+q2)*(Vs-eatp) + g34*(q3+q4)*(Vs-eatp))
    I_ds = gcan*(Vd-Vs)/k # modification of somatic current due to dendritic current
    # currents in Dendrite
    I_can = gcan*caninf*(Vd-ena) # calcium current
    I_sd = gcan*(Vs-Vd)/(1-k) # modification of dendritic current due to somatic current
    
    if collect_current_data:
        atp_current.append(I_atp)
        ca_current.append(I_can)
        nap_current.append(I_naps)
        na_current.append(I_nas)
        k_current.append(I_ks)
        timeArray.append(t)
    
    #### DIFFERENTIAL EQUATIONS
    # SOMATIC EQUATIONS
    dVs = (-I_ks-I_nas-I_naps-I_L-I_sd-I_atp)/Cms
    dns = (ninfs-ns)/tauns
    dhs = (hinfs-hs)/tauhs
    # DENDRITIC EQUATIONS
    dVd = (-I_can-I_ds)/Cmd
    dC = fi/Vi*( J_ER_in - J_ER_out)
    dl = A*( Kd - (C + Kd)*l )
    #P2X7R EQUATIONS
    # states
    dq1 = 2.*k4*atp*c2 + 3.*k5*q2 - (2.*k3 + k6*atp)*q1
    dq2 = k6*atp*q1 + L2*q3 - (3.*k5 + L3)*q2
    dq3 = k2*atp*q4 + L3*q2 -(3.*k1 + L2)*q3
    dq4 = 2.*k2*atp*c3 + 3.*k1*q3 - (2.*k1 + k2*atp)*q4
    dc1 = k1*c2 + L1*c4 - 3.*k2*atp*c1
    dc2 = 3.*k2*atp*c1 + 2.*k3*q1 - (k1+2.*k4*atp)*c2
    dc3 = 3*k2*atp*c4 + 2.*k1*q4 - (k1 + 2.*k2*atp)*c3
    dc4 = k1*c3 - (L1 + 3.*k2*atp)*c4
    # atp concentration
    datp = atpFunct(atp, t)

    dy = [dVs,dns,dhs,dVd,dC,dl,dq1,dq2,dq3,dq4,dc1,dc2,dc3,dc4,datp]     
    return dy

def df_Ca(y, t, Iaps , gnaps , gk , vmp , gL , Ka , Ke , Kd , Ki , Cmd , eSyn , P , \
    Cms , taunb , gna , Kcan , Ct , ek , vh , vm , vn , gtonic , eL , LL , tauhb , \
    smp , A , gcan , ena , I, fi , ncan , Ve , Vi , k , \
    sh , sn , sm , sigma):

    Vs = y[0]
    ns = y[1]
    hs = y[2]
    Vd = y[3] #Dendritic membrane voltage
    C = y[4] #Intracellular Ca2+ balance
    l = y[5] #IP3 channel gating variable

    # SOMATIC FUNCTIONS
    minfs = 1/(1+math.exp((Vs-vm) /sm))
    ninfs = 1/(1+math.exp((Vs-vn) /sn))
    minfps = 1/(1+math.exp((Vs-vmp)/smp))
    hinfs = 1/(1+math.exp((Vs-vh) /sh))

    tauns = taunb/math.cosh((Vs-vn)/(2*sn))
    tauhs = tauhb/math.cosh((Vs-vh)/(2*sh))

    # DENDRITIC FUNCTIONS
    #Calculate ER Ca
    Ce = (Ct - C)/sigma
    # Flux of Ca from ER to cytosol(regulated by IP3 receptors)
    J_ER_in=(LL + P*math.pow(((I*C*l)/((I+Ki)*(C+Ka))),3))*(Ce - C) 
    # Flux from cytosol back to ER (controlled by SERCA pumps)
    J_ER_out=Ve*math.pow(C,2)/(math.pow(Ke,2)+math.pow(C,2))
    # Activation of calcium current (I_can)
    caninf = 1/(1+math.pow((Kcan/C),ncan))

    # CURRENT EXPRESSIONS
    # currents in Soma
    I_nas = gna*math.pow(minfs,3)*(1-ns)*(Vs-ena) #sodium current
    I_ks = gk*math.pow(ns,4)*(Vs-ek) #potassium current
    I_naps = gnaps*minfps*hs*(Vs-ena) #persistent sodium current
    I_L = gL*(Vs-eL) #pA
    I_ds = gcan*(Vd-Vs)/k # modification of somatic current due to dendritic current
    # currents in Dendrite
    I_can = gcan*caninf*(Vd-ena) # calcium current
    I_sd = gcan*(Vs-Vd)/(1-k) # modification of dendritic current due to somatic current
    
    #### DIFFERENTIAL EQUATIONS
    # SOMATIC EQUATIONS
    dVs = (-I_ks-I_nas-I_naps-I_L-I_sd)/Cms
    dns = (ninfs-ns)/tauns
    dhs = (hinfs-hs)/tauhs
    # DENDRITIC EQUATIONS
    dVd = (-I_can-I_ds)/Cmd
    dC = fi/Vi*( J_ER_in - J_ER_out)
    dl = A*( Kd - (C + Kd)*l )

    dy = [dVs,dns,dhs,dVd,dC,dl]     
    return dy

def zero_slope(atp, t):
    return 0.0

def printOutput(data, path):
    try:
        data.to_csv(path+'.csv')
        print "Printing ", path+'.csv'
    except:
        print "Could not print ", path+'.csv'
        
def timer(label,start):
    if timing_on:
        print label, ": ",clock() - start
    
def solve(df, t, atpFunct, Iaps , gnaps , gk , vmp , gL , Ka , Ke , Kd , Ki , Cmd , eSyn , P , Cms , taunb , g34 , \
          gna , Kcan , Ct , ek , vh , vm , vn , gtonic , eL , LL , tauhb , smp , k3 , k2 , k1 , k6 , k5 , k4 , \
          A , gcan , ena , I ,L2 , L3 , L1 , fi , ncan , Ve , Vi , k , eatp ,sh , sn , sm , sigma , g12, \
          Vs0, ns0, hs0, Vd0, C0, l0, q1o, q2o, q3o, q4o, c1o, c2o, c3o, c4o, atp0):
    '''
    Pass constants to this function by calling solve(df, t, f_atp, **parameter_dict)
    Unpacks the parameters dictionary, makes y0 array, makes the time array, calls odeint, and passes parameters to odeint.
    '''
    
    #only works for df_atp and df_Ca b/c I can't pass all the args necessary for df_atp into df_Ca or it won't work
    #(too many args passed into a function.) Therefore I can't have a single, general odeint function call.
    s = clock()
    if df == df_atp:
    #solve df, the retrieve Vs from what odeint returns (odeint_solution.T[0] since Vs is the first column)
        y = spi.odeint(df, [Vs0, ns0, hs0, Vd0, C0, l0, q1o, q2o, q3o, q4o, c1o, c2o, c3o, c4o, atp0], t, \
                    args=(atpFunct, Iaps , gnaps , gk , vmp , gL , Ka , Ke , Kd , Ki , \
                                        Cmd , eSyn , P , Cms , taunb , g34 , gna , Kcan , Ct , ek , vh , \
                                        vm , vn , gtonic , eL , LL , tauhb , smp , k3 , k2 , k1 , k6 , k5 , \
                                        k4 , A , gcan , ena , I , L2 , L3 , L1 , fi , ncan , Ve , Vi , \
                                        k , eatp , sh , sn , sm , sigma , g12)).T
    elif df == df_Ca:
        y = spi.odeint(df, [Vs0, ns0, hs0, Vd0, C0, l0], t, \
                    args=(Iaps , gnaps , gk , vmp , gL , Ka , Ke , Kd , Ki , Cmd , eSyn , P , \
                          Cms , taunb , gna , Kcan , Ct , ek , vh , vm , vn , gtonic , eL , LL , tauhb , \
                          smp , A , gcan , ena , I, fi , ncan , Ve , Vi , k , \
                          sh , sn , sm , sigma)).T
    else:
        print "Parameter df must be either df_Ca or df_atp."
        return None
    timer(' odeint',s)
    return y[0]


def change_constant(var_name, var_vals, df, params, atpFunc = (lambda x,t: 0.0)):
    
    const_params = copy.deepcopy(params)
    time = np.linspace(0.0, tEnd, tEnd/dt)
    #set initial conditions
    v_array = []
    if var_name in const_params:
        for val in var_vals:
            const_params[var_name] = val
            print var_name, "  ", val, "  "
            sys.stdout.flush()  
            s_odeint = clock()
            v = solve(df, time, atpFunc, **const_params)
            print v
            v_array.append(pd.Series(data = v, name = str(val), index = time))
            timer(" unpack + odeint", s_odeint)
            
    else:
        raise KeyError("There is no parameter with the symbol {}. Possibly parameter symbols are:"\
                            "{}.".format(var_name,str(const_params.keys()).strip('[]')))
    d_atp = pd.concat(v_array, axis=1)
    return d_atp
````

````python
#parallelized version
from IPython.parallel import Client

c = Client()

with c[:].sync_imports():
    import scipy.integrate as spi
    import numpy as np
    #import warnings
    import pandas as pd
    #import math
    import copy

lbv = c.load_balanced_view()
lbv.block = True

def oneVal(var_name, val, params, time):
    
    #copy params so that I can be sure that each parallel process is independent (is this necessary?)
    const_params = copy.deepcopy(params)
    const_params[var_name] = val
    v = solve(df, time, atpFunc, **const_params)
    s = pd.Series(data = v, name = str(val), index = time)
    v_array.append(s)
    return

def change_constant(var_name, var_vals, df, params, atpFunc = (lambda x,t: 0.0)):
    
    time = np.linspace(0.0, tEnd, tEnd/dt)
    #set initial conditions
    v_array
    if var_name in params:
        #processes within the for loop can be done in parallel, rather than serially
        v_array = lbv.map(lambda x: oneVal(var_name,x,params,time),var_vals)            
            
    else:
        raise KeyError("There is no parameter with the symbol {}. Possibly parameter symbols are:"\
                            "{}.".format(var_name,str(const_params.keys()).strip('[]')))

    d_atp = pd.concat(v_array, axis=1)
    return d_atp
````

````python
def unpack(df, t, atpFunct, Iaps , gnaps , gk , vmp , gL , Ka , Ke , Kd , Ki , Cmd , eSyn , P , Cms , taunb , g34 , \
          gna , Kcan , Ct , ek , vh , vm , vn , gtonic , eL , LL , tauhb , smp , k3 , k2 , k1 , k6 , k5 , k4 , \
          A , gcan , ena , I ,L2 , L3 , L1 , fi , ncan , Ve , Vi , k , eatp ,sh , sn , sm , sigma , g12, \
          Vs0, ns0, hs0, Vd0, C0, l0, q1o, q2o, q3o, q4o, c1o, c2o, c3o, c4o, atp0):

    y = spi.odeint(df, [Vs0, ns0, hs0, Vd0, C0, l0, q1o, q2o, q3o, q4o, c1o, c2o, c3o, c4o, atp0], t, \
                    args=(atpFunct, Iaps , gnaps , gk , vmp , gL , Ka , Ke , Kd , Ki , \
                                        Cmd , eSyn , P , Cms , taunb , g34 , gna , Kcan , Ct , ek , vh , \
                                        vm , vn , gtonic , eL , LL , tauhb , smp , k3 , k2 , k1 , k6 , k5 , \
                                        k4 , A , gcan , ena , I , L2 , L3 , L1 , fi , ncan , Ve , Vi , \
                                        k , eatp , sh , sn , sm , sigma , g12)).T
    return y[0], y[6:15]

def atpf(atp, t):
    washout_delay = 5 #ms
    slope = -param['atp0']/washout_delay
    if t < 1000*40:
        return 0.0
    elif atp >= 0.0:
        return slope*t
    else:
        return 0.0
````

````python
def df(y, t, atpFunct):
    
    Vs = y[0]
    ns = y[1]
    hs = y[2]
    Vd = y[3] #Dendritic membrane voltage
    C = y[4] #Intracellular Ca2+ balance
    l = y[5] #IP3 channel gating variable
    q1 = y[6]
    q2 = y[7]
    q3 = y[8]
    q4 = y[9]
    c1 = y[10]
    c2 = y[11]
    c3 = y[12]
    c4 = y[13]
    atp = y[14]

    # SOMATIC FUNCTIONS
    minfs = 1/(1+math.exp((Vs-vm) /sm))
    ninfs = 1/(1+math.exp((Vs-vn) /sn))
    minfps = 1/(1+math.exp((Vs-vmp)/smp))
    hinfs = 1/(1+math.exp((Vs-vh) /sh))

    tauns = taunb/math.cosh((Vs-vn)/(2*sn))
    tauhs = tauhb/math.cosh((Vs-vh)/(2*sh))

    # DENDRITIC FUNCTIONS
    #Calculate ER Ca
    Ce = (Ct - C)/sigma
    # Flux of Ca from ER to cytosol(regulated by IP3 receptors)
    J_ER_in=(LL + P*math.pow(((I*C*l)/((I+Ki)*(C+Ka))),3))*(Ce - C) 
    # Flux from cytosol back to ER (controlled by SERCA pumps)
    J_ER_out=Ve*math.pow(C,2)/(math.pow(Ke,2)+math.pow(C,2))
    # Activation of calcium current (I_can)
    caninf = 1/(1+math.pow((Kcan/C),ncan))

    # CURRENT EXPRESSIONS
    # currents in Soma
    I_nas = gna*math.pow(minfs,3)*(1-ns)*(Vs-ena) #sodium current
    I_ks = gk*math.pow(ns,4)*(Vs-ek) #potassium current
    I_naps = gnaps*minfps*hs*(Vs-ena) #persistent sodium current
    I_L = gL*(Vs-eL) #pA
    I_atp = g12*(q1+q2)*(1/10)*(Vs-eatp) + g34*(q3+q4)*(Vs-eatp)
    I_ds = gcan*(Vd-Vs)/k # modification of somatic current due to dendritic current
    # currents in Dendrite
    I_can = gcan*caninf*(Vd-ena) # calcium current
    I_sd = gcan*(Vs-Vd)/(1-k) # modification of dendritic current due to somatic current

    #### DIFFERENTIAL EQUATIONS
    # SOMATIC EQUATIONS
    dVs = (-I_ks-I_nas-I_naps-I_L-I_sd-I_atp)/Cms
    dns = (ninfs-ns)/tauns
    dhs = (hinfs-hs)/tauhs
    # DENDRITIC EQUATIONS
    dVd = (-I_can-I_ds)/Cmd
    dC = fi/Vi*( J_ER_in - J_ER_out)
    dl = A*( Kd - (C + Kd)*l )
    #P2X7R EQUATIONS
    # states
    dq1 = 2.*k4*atp*c2 + 3.*k5*q2 - (2.*k3 + k6*atp)*q1
    dq2 = k6*atp*q1 + L2*q3 - (3.*k5 + L3)*q2
    dq3 = k2*atp*q4 + L3*q2 -(3.*k1 + L2)*q3
    dq4 = 2.*k2*atp*c3 + 3.*k1*q3 - (2.*k1 + k2*atp)*q4
    dc1 = k1*c2 + L1*c4 - 3.*k2*atp*c1
    dc2 = 3.*k2*atp*c1 + 2.*k3*q1 - (k1+2.*k4*atp)*c2
    dc3 = 3*k2*atp*c4 + 2.*k1*q4 - (k1 + 2.*k2*atp)*c3
    dc4 = k1*c3 - (L1 + 3.*k2*atp)*c4
    # atp concentration
    datp = atpFunc(atp, t)

    dy = [dVs,dns,dhs,dVd,dC,dl,dq1,dq2,dq3,dq4,dc1,dc2,dc3,dc4,datp]     
    return dy
````

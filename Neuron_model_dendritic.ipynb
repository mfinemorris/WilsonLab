```python
'''
In the following, m, h, and n are gating variables controlling the activation or inactivation of certain channels.
M is approximated as starting instantaneously, so we never need a dm/dt equation-we can just use m_inf in place of m.

This version has the dendritic component included, taken from the 2011 paper by Toparikova and Butera.
One assumption made is that total Ca^(2+) in cell is const, so we can neglect flux of Ca 2+ into or out of cell. 
Therefore, Ca_ER = (Ca_tot - Ca_i)/sig, where Ca_tot = 1.25 uM and sig = 0.185. (sig is ratio of cytosolic to ER vol.)


Helpful resources for developing this:
Code examples:
http://www.briansimulator.org/docs/examples-frompapers_Rothman_Manis_2003.html
http://www.briansimulator.org/docs/examples-misc_cable.html?highlight=dendrit
http://www.briansimulator.org/docs/librarymodels.html?highlight=dendrit
http://www.briansimulator.org/docs/examples-misc_using_classes.html

http://www.briansimulator.org/docs/examples-frompapers_Kremer_et_al_2011.html

Documents:
http://fog.ccsf.cc.ca.us/~wkaufmyn/ENGN45/Course%20Handouts/15_ElectricalProps/05_TemperatureConductivityConductor.html


'''


#number of seconds to run simulation (unitless)
runLen = 2

from brian import *
from brian.library.ionic_currents import *
import time
import inspect
from brian.compartments import *
#%matplotlib inline


#microMolar
uM = (10**-3)*mole/meter**3
#picoLitre
pL = (10**3)*(umeter**3)
#attomole
aMol = (10**-18)*mole

#start the run time clock
start = time.clock()
print "Start: ", time.strftime("%H:%M:%S")
def checkpoint():
    ellapsed = (time.clock() - start)
    print "{} sec ellapsed at {}".format(ellapsed, time.strftime("%H:%M:%S"))
    
#reinit the simulator clock so that the simulator starts at 0 seconds
defaultclock.reinit()

#---Model from 2011 Toparikova and Butera paper

EL = -57 * mvolt
EK = -85.0*mvolt
ENa = 50.0*mvolt

Cs = 21 * uF
Cd = 5 * uF

gL = 2.3 * nsiemens
gC = 1.0 * nsiemens
gK = 11.2 * nsiemens
gNa = 28.0 * nsiemens
gNaP = 2.0 * nsiemens #or 1.5 * nsiemens
gCaN = 1.5 * nsiemens

#constants for K and Na
thetaN = -29*mvolt
thetaM = -34*mvolt
sigmaN = -4*mvolt
sigmaM = -5*mvolt

#constants for NaP and Nap-h
thetaH = -48*mvolt
sigmaH = 6*mvolt
thetaP = -40*mvolt
sigmaP = -6*mvolt

#eqns for Na and K
mInf = lambda v: (1+exp((v-thetaM)/sigmaM))**-1
nInf = lambda v: (1+exp((v-thetaN)/sigmaN))**-1
tauMaxN = 10*ms
tauN = lambda v: tauMaxN/cosh((v-thetaN)/(2*sigmaN))

#eqns for NaP and NaP-h
hInf = lambda v: (1+exp((v-thetaH)/sigmaH))**-1
pInf = lambda v: (1+exp((v-thetaP)/sigmaP))**-1
tauMaxH = 10000*ms
tauH = lambda v: tauMaxH/cosh((v-thetaH)/(2*sigmaH))


#for calcium kinetics
k = 0.3
#funct. of Ca describing activation of ICaN (calcium current) by Ca2+. Called f([Ca]_{i}) in paper
ICaN_active = lambda Ca: (1+(K_CaN/Ca)**nc)**-1
nc = 0.97
K_CaN = 0.74 * uM

fi = 0.000025
A = 0.005

L_IP3R = 0.37 * pL/second #ER Leak permiability
P_IP3R = 31000 * pL/second
IP3 = 1 * uM
#dissociation consts. for IP3 receptor activation by IP3 and Ca2+ respectively
K_I= 1.0 * uM
K_a = 0.4 * uM
K_d = 0.4 * uM
VSERCA = 400 * aMol/second #maximal SERCA pump rate
KSERCA = 0.2 * uM #coefficient for SERCA pumps
#for calc [Ca]_{ER}
Ca_tot = 1.25*uM
sig = 0.185 #ratio of cytosolic to ER volume

eqs = Equations('''
dVs/dt = -(INa + INaPh + IK + Ileak + (Vs-Vd)*gC/(1-k))/Cs:mvolt
dVd/dt = -(ICaN + (Vd-Vs)*gC/k)/Cd:mvolt
#Currents
INa = (gNa*(1-n)*(Vs-ENa)*m_inf**3):amp
INaPh = (gNaP*(Vs-ENa)*h*p_inf):amp
IK = (gK*(Vs-EK)*n**4):amp
Ileak = gL*(Vs-EL):amp
ICaN = gCaN*ICaN_active(Ca_i)*(Vd-ENa):amp
#Sodium and Potassium Activation
m_inf = mInf(Vs):1
n_inf = nInf(Vs):1
tau_n = tauN(Vs):ms
dn/dt =  (n_inf - n)/tau_n:1
#
#sodium inactivation (nap or naph)
p_inf = pInf(Vs):1
h_inf = hInf(Vs):1
tau_h = tauH(Vs):ms
dh/dt = (h_inf - h)/tau_h:1
I:amp
#calcium kinetics
dCa_i/dt = fi*(J_ERin-J_ERout):uM
dl/dt = A*(K_d - l*(Ca_i+K_d)):1
J_ERin = (L_IP3R + P_IP3R*(IP3*Ca_i*l/((IP3+K_I)*(Ca_i+K_a)))**3)(((Ca_tot - Ca_i)/sig) - Ca_i):aMol/second
J_ERout = VSERCA*(Ca_i**2/(KSERCA**2 + Ca_i**2)):aMol/second
''')
```

```python
from brian import *
from brian.library.ionic_currents import *
from brian.compartments import *
%matplotlib inline

#---from 2011 Toporikova paper
#microMolar
uM = (10**-3)*mole/meter**3
#picoLitre
pL = (10**3)*(umeter**3)
#attomole
aMol = (10**-18)*mole

EL = -57*mvolt
EK = -85.0*mvolt
ENa = 50.0*mvolt
C = 21 * uF
Cd = 5 * uF
gL = 2.3*nsiemens
gK = 11.2*nsiemens
gNa = 28*nsiemens
gNaP = 2.0*nsiemens
gC = 1.0 * nsiemens
k = 0.3

#---Model 1 of butera99a
#The nominal params for leak current are gL=2.8 nS and EL=-65 mV.

EL = -57*mvolt
EK = -85.0*mvolt
ENa = 50.0*mvolt
C = 21 * pF
gL = 2.8*nsiemens
gK = 11.2*nsiemens
gNa = 28*nsiemens
gNaP = 2.8*nsiemens

#time consts
#---time constants are consistent for all currents in Model 1 of butera99a
tauMaxH = 10000*ms
tauMaxN = 10*ms

#constants for K and Na
thetaN = -29*mvolt
thetaM = -34*mvolt
sigmaN = -4*mvolt
sigmaM = -5*mvolt

#constants for NaP and Nap-h
thetaH = -48*mvolt
sigmaH = 6*mvolt
thetaP = -40*mvolt
sigmaP = -6*mvolt

#eqns for Na and K
mInf = lambda v: (1+exp((v-thetaM)/sigmaM))**-1
nInf = lambda v: (1+exp((v-thetaN)/sigmaN))**-1
tauN = lambda v: tauMaxN/cosh((v-thetaN)/(2*sigmaN))

#eqns for NaP and NaP-h
hInf = lambda v: (1+exp((v-thetaH)/sigmaH))**-1
pInf = lambda v: (1+exp((v-thetaP)/sigmaP))**-1
tauH = lambda v: tauMaxH/cosh((v-thetaH)/(2*sigmaH))


def makeSoma():
    multiplierS = k
    #soma equation
    soma = MembraneEquation(C*multiplierS) + IonicCurrent('ILeak=multiplierS*gL*(vm-EL) : amp')
    soma += IonicCurrent('INa=multiplierS*(gNa*(1-n)*(vm-ENa)*m_inf**3) : amp')
    soma += IonicCurrent('INaPh=multiplierS*(gNaP*(vm-ENa)*h*p_inf) : amp')
    soma += IonicCurrent('IK=multiplierS*(gK*(vm-EK)*n**4) : amp')
    soma += Equations(''' 
    #Sodium and Potassium Activation
    m_inf = mInf(vm):1
    n_inf = nInf(vm):1
    tau_n = tauN(vm):ms
    dn/dt =  (n_inf - n)/tau_n:1
    #sodium inactivation (nap or naph)
    p_inf = pInf(vm):1
    h_inf = hInf(vm):1
    tau_h = tauH(vm):ms
    dh/dt = (h_inf - h)/tau_h:1
    I:amp
    ''')
    return soma

def makeDendrite():
    #for calcium kinetics
    multiplierD = (k**2)/(1-k)
    #funct. of Ca describing activation of ICaN (calcium current) by Ca2+. Called f([Ca]_{i}) in paper
    ICaN_active = lambda Ca: (1+(K_CaN/Ca)**nc)**-1
    gCaN = 1.5
    nc = 0.97
    K_CaN = 0.74
    fi = 0.000025
    A = 0.005
    L_IP3R = 0.37 #ER Leak permiability
    P_IP3R = 31000
    IP3 = 1
    #dissociation consts. for IP3 receptor activation by IP3 and Ca2+ respectively
    K_I= 1.0
    K_a = 0.4
    K_d = 0.4
    VSERCA = 400 #maximal SERCA pump rate
    KSERCA = 0.2 #coefficient for SERCA pumps
    #for calc [Ca]_{ER}
    Ca_tot = 1.25
    sig = 0.185 #ratio of cytosolic to ER volume

    dendrite = MembraneEquation(Cd*multiplierD)
    dendrite += IonicCurrent('ICaN=multiplierD*gCaN*ICaN_active(Ca_i)*(vm-ENa)*amp : amp')
    dendrite += Equations(''' 
    #calcium kinetics
    dCa_i/dt = fi*(J_ERin-J_ERout)*uM:uM
    dl/dt = A*(K_d - l*(Ca_i+K_d)):1
    J_ERin = (L_IP3R + P_IP3R*(IP3*Ca_i*l/((IP3+K_I)*(Ca_i+K_a)))**3)(Ca_ER - Ca_i)*(aMol/second):aMol/second
    J_ERout = VSERCA*(Ca_i**2/(KSERCA**2 + Ca_i**2))*(aMol/second):aMol/second
    Ca_ER = uM*(Ca_tot - Ca_i)/sig:uM
    I:amp
    ''')
    return dendrite

dendrite = makeDendrite()
soma = makeSoma()
print dendrite

import sys
sys.stdout.flush()

neuron_eqs=Compartments({'soma':soma, 'dendrite':dendrite})
neuron_eqs.connect('soma','dendrite',k/(gC*(k-1)))
print neuron_eqs
sys.stdout.flush()

#reinit the simulator clock so that the simulator starts at 0 second
defaultclock.reinit()
#when ready set 2nd arg to neuron_eqs
neurons=NeuronGroup(1, model=neuron_eqs, order = 2, clock = defaultclock)

sys.exit(0)
neurons.vm_soma = EL
neurons.h = hInf(EL)
neurons.n = nInf(EL)
print 
M = StateMonitor(neurons,"vm_soma", record=True)
M = StateMonitor(neurons,"vm_dendrite", record=True)
run(1*second)
#checkpoint()
pd=plot(M.times/ms, M[0]/mvolt)
#xlabel('Time (ms)')
#ylabel('Membrane potential (mV)')
plt.show()
```
